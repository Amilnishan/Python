```python
#1.  Create modules Rectangle, Circle. Include methods to find area and perimeter of respective figures in each module. Write
#programs that finds area and perimeter of figures by different importing statements

import rectangle
import circle

while(1) :
    print("1.Rectangle \n2.Circle \n3.Exit")
    ch=int(input("enter your choice:"))
    if ch == 1 :
        l=int(input("enter the length:"))
        b=int(input("enter the breadth:"))
        print("Area of rectangle =",rectangle.rectarea(l,b))
        print("Perimeter of rectangle =",rectangle.rectperi(l,b))
    elif ch == 2 :
        r=int(input("Enter the radius:"))
        print("Area of circle =",circle.circlearea(r))
        print("Perimeter of circle =",circle.circleperi(r))
    else :
        print("Exiting...")
        break;
```

```python
"""2. Create a package graphics with modules rectangle, circle and sub package graphics3D with modules cuboid and sphere. Include
methods to find area and perimeter of respective figures in each module. Write programs that finds area and perimeter of figures by
different importing statements. (Include selective import of modules and import * statements)"""

#main.py

from graphics import graphics_rect
from graphics import graphics_circle
from graphics.graphics3D import cuboid
from graphics.graphics3D import sphere


while(1):
    print("\n 1.Rectangle \n 2.Circle \n 3.Cuboid \n 4.Sphere \n 5.Exit")
    ch=int(input("Enter your choice:"))
    if ch == 1 :
        l=int(input("Enter the length:"))
        b=int(input("Enter the breadth:"))
        graphics_rect.rectarea(l,b)
        graphics_rect.rectperi(l,b)
    elif ch == 2 :
        r=int(input("Enter the radius:"))
        graphics_circle.circlearea(r)
        graphics_circle.circleperi(r)
    elif ch == 3 :
        l=int(input("Enter the length:"))
        b=int(input("Enter the breadth:"))
        h=int(input("Enter the height:"))
        cuboid.cuboidarea(l,b,h)
        cuboid.cuboidperi(l,b,h)
    elif ch == 4 :
        r=int(input("Enter the radius:"))
        sphere.spherearea(r)
        sphere.sphereperi(r)
    else :
        print("Exiting...")
        break
```
